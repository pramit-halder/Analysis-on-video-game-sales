# Importing dependencies

# Loading the dataset

```python
data = pd.read_csv("vgsales.csv")
```

```python
data
```

# Data Cleaning

## 1. Removing null values

```python
data.info()
```

```python
pd.isnull(data).sum(axis=0)
```

```python
data.dropna(inplace=True)
```

```python
pd.isnull(data).sum(axis=0)
```

## 2. Converting the data type where required

```python
data["Year"]=data["Year"].astype("int64")
```

```python
data.sample(10)
```

# Plots

## 1. Platform

```python
sb.set(rc={"figure.figsize":(20,20)})
plat_count = sb.countplot(x="Platform",data=data)
for bar in plat_count.containers:
    plat_count.bar_label(bar)
```

### So, more number of games are released for DS.

# Predicting the global sales on the basis of sales in NA,EU,JP and others.

```python
df1 = data.drop(columns = ["Rank","Name","Platform","Year","Genre","Publisher"])
df1.head()
```

```python
X = df1.iloc[:,:4]
y = df1.iloc[:,-1]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
acc = model.score(X_test,y_test)
print(f"The coeficients are :- {model.coef_}")
print(f"The intercept is :- {model.intercept_}")
print(acc)
```

```python
from sklearn.metrics import r2_score
y_train_pred = model.predict(X_train)
acc_train = r2_score(y_train_pred,y_train)
acc_train
```

```python
y_test_pred = model.predict(X_test)
acc_test = r2_score(y_test,y_test_pred)
acc_test
```
